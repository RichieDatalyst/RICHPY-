**Smoothing**

```python
def laplace_smoothing(tokens):
    # Create a frequency dictionary for words in tokens
    word_counts = defaultdict(int)
    for word in tokens:
        word_counts[word] += 1

    # Define the vocabulary size (number of unique words)
    vocabulary_size = len(word_counts)

    # Add 1 to the count of each word in the vocabulary
    smoothed_word_counts = {word: count + 1 for word, count in word_counts.items()}

    # Calculate probabilities for each word
    word_probabilities = {word: count / (len(tokens) + vocabulary_size) for word, count in smoothed_word_counts.items()}
    return word_probabilities

    # add-one smoothing to ensure that no word has a zero probability of occurring.
```

```python
X_train_smoothed_probabilities = X_train_without_stopwords.apply(laplace_smoothing)
print(X_train_smoothed_probabilities)
```

**Logistic Regression**

```python
class Logistic_Regression():
    def __init__(self, learning_rate, iter_num):
        self.learning_rate = learning_rate
        self.iter_num = iter_num
        self.weight = None
        self.bias = None

    def fit(self, X, y):
        self.rows, self.columns = X.shape
        self.weight = np.zeros(self.columns)
        self.bias = 0
        self.X = X
        self.y = y
        for i in range(self.iter_num):
            self.update_weights()

    def update_weights(self):
        Y_hat = 1 / (1 + np.exp(- (self.X.dot(self.weight) + self.bias)))
        dw = (1 / self.rows) * np.dot(self.X.T, (Y_hat - self.y))
        db = (1 / self.rows) * np.sum(Y_hat - self.y)
        self.weight = self.weight - self.learning_rate * dw
        self.bias = self.bias - self.learning_rate * db

    def predict(self, X):
        Y_pred = 1 / (1 + np.exp(- (X.dot(self.weight) + self.bias)))
        Y_pred = np.where(Y_pred > 0.5, 1, 0)
        return Y_pred
```

```python
classifier_bow = Logistic_Regression(learning_rate = 0.01, iter_num = 1000)
```

```python
classifier_tfidf = Logistic_Regression(learning_rate = 0.01, iter_num = 1000)
```

```python
bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)
classifier_bow.fit(X_train_bow.toarray(), y_train) # training logistic regression model using bow representation
```

```python
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
classifier_tfidf.fit(X_train_tfidf.toarray(), y_train) # training logistic regression model using tfidf representation
```
